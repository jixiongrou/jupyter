```python
#数据读取
df_sheet_1 = pd.read_excel(io='test.xlsx', sheet_name="Sheet1")
df_sheet_2 = pd.read_excel(io='test.xlsx', sheet_name="Sheet2")
df_sheet_3 = pd.read_excel(io='test.xlsx', sheet_name="Sheet3")
```

```python
#数据清洗
df_sheet_1.drop_duplicates(inplace=True)    #清除重复数据
df_sheet_1.dropna(axis=0, how='any', inplace=True)    #删除包含空字段的行,axis=0:按行,how=any:有空就删,
df_sheet_1 = df_sheet_1.reset_index(drop=True)    #重新排index
```

```python
#数据合并
df_sheet_1 = df_sheet_1.loc[0:229]
```

```python
dataframes = [df_sheet_1, df_sheet_2]
df_tmp = pd.concat(dataframes)
```

```python
df_tmp = df_tmp.reset_index(drop=True)
df_tmp
```

```python
df_result = pd.merge(df_tmp, df_sheet_3)
```
